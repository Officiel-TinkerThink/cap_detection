```python
import cv2
import os
from pathlib import Path
import pandas as pd
```

```python
root = Path(os.getcwd()).parent
root
```

```python
from pathlib import Path

sample_path = Path(root) / 'sample'

samples = []
# get list of file from that sample path
for file in sample_path.iterdir():
  if not file.stem in samples:
    samples.append(file.stem)

print(samples)
```

```python
df = pd.DataFrame(columns=['sample', 'label', 'x', 'y', 'w', 'h'])
df
```

```python
data = []
for sample in samples:
  with open(Path(sample_path/ f"{sample}.txt"), 'r', encoding='utf8') as file:
    all_lines = file.readlines()
    for line in all_lines:
      items = line.split()
      data.append({
        'sample':sample,
        'label':items[0],
        'x':items[1],
        'y':items[1],
        'w':items[1],
        'h':items[1],
      })

df = pd.DataFrame(data)
      
```
